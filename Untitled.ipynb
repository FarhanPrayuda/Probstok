# Menyimpulkan ECDF dengan Left Tail

```python
def ecdf(data):
    xaxis = np.sort(data)
    yaxis = np.arange(1,len(data)+1)/len(data)
    return xaxis, yaxis
x,y = ecdf(df['AveragePrice'])
```

```python
plt.plot(x,y,linestyle='none',marker='.')
plt.title('ECDF of Average Stock Prices')
plt.xlabel('Price per Unit')
plt.ylabel('Percentage')
plt.margins(0.01)
plt.show()
```

```python
def plot_ecdf(data,title = "ECDF Plot", xlabel = 'Data Values', ylabel = 'Percentage'):
    
    """ 
    Function to plot ecdf taking a column of data as input.
    """
    xaxis = np.sort(data)
    yaxis = np.arange(1, len(data)+1)/len(data)
    plt.plot(xaxis,yaxis,linestyle='none',marker='.')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.margins(0.02)
```

```python
plot_ecdf(df['AveragePrice'])
plt.show()
```
